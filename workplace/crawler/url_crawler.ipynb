### Get all URL of players in FIFA22, England Premier League (1)

```python
# Open the fifa index website
import urllib.request as req
import bs4
import re

home = "https://www.fifaindex.com"
init_url = "https://www.fifaindex.com/players/"
with req.urlopen(init_url) as response:
    data = response.read().decode("utf-8")

root=bs4.BeautifulSoup(data, "html.parser")
```

```python
%%capture

# Get all date urls
total_urls = root.find_all("a", {'class' : 'dropdown-item', 'href' : re.compile("^/players/")})

# Get urls of date btn
date_urls = total_urls[20:]

for idx, row in enumerate(date_urls):
    date_urls[idx] = home + date_urls[idx]["href"]

"""
date_urls = 
['https://www.fifaindex.com/players/',
 'https://www.fifaindex.com/players/fifa22_534/',
 'https://www.fifaindex.com/players/fifa22_533/',
 'https://www.fifaindex.com/players/fifa22_532/',
.
.
.]

"""
```

```python
# All the needed league urls
league_urls = [ "?gender=0&league=13&order=desc", 
                "?gender=0&league=16&order=desc", 
                "?gender=0&league=19&order=desc", 
                "?gender=0&league=31&order=desc", 
                "?gender=0&league=53&order=desc", 
                "?gender=0&league=308&order=desc" ]

league_url = ["?gender=0&league=13&order=desc"]
```

```python
import requests

# Get all involved players' urls in FIFA22
player_urls = [] 
player_ids = []

for date in date_urls :                                  # = https://www.fifaindex.com/players/fifa22_534/
    for league in league_url:                           # = ?gender=0&league=13&order=desc
            # get the whole url
            current_url = date + league

            # connect to the url
            with req.urlopen(current_url) as response:
                data = response.read().decode("utf-8")

            cur=bs4.BeautifulSoup(data, "html.parser")

            # get the num of last page
            last_page = cur.find_all("a", string=re.compile("Last"))[0]["href"]
            last_page = last_page.split("=",1)[1] 
            last_page = last_page.split("&")[0]
            last_page = int(last_page)

            # crawl the page from 1 to its lastpage(ex: 577)
            for page in range(1, last_page+1):
                # 連結網站
                response = requests.get(date + "?page=" + str(page) + "&" + league[1:])  # = https://www.fifaindex.com/players/?page=1gender=0&league=13&order=desc
                print("\n currently on: ", date + "?page=" + str(page) + "&" + league[1:])

                # HTML原始碼解析
                soup = bs4.BeautifulSoup(response.text, "html.parser")

                # 取得所有href開頭為/player/的<a>(所有player的詳細資料), note: 會有重複的重況
                found = soup.find_all("a", {'href' : re.compile("^/player/")})

                print(f"====================第{str(page)}頁====================")
                for idx, url in enumerate(found):
                    found[idx] = home + url["href"]
                    replaced = found[idx].replace('fifa22/', '')
                    player_id = re.findall(r'\d+', replaced)[0]
                    if( player_id not in player_ids): # check if it's already in list
                        print(replaced, player_id)  # 取得連結
                        player_ids.append(player_id)
                        player_urls.append(replaced)

```

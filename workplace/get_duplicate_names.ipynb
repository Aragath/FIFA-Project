```python
import pandas as pd

DATA_PATH = "pre-processed/player_stats.csv"
df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df1 = df[['name', 'ID']]

df1
```

```python
df2 = df1[df1.groupby(['name'])['ID'].transform('nunique') > 1].drop_duplicates()

df2
```

```python
df2 = df2.sort_values(by=["name"]) 
df2
```

```python
header = df2.columns

df2.to_csv('datasets/duplicate_names.csv', index = False, header = header, encoding='utf-8-sig')
```

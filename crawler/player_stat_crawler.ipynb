```python
# change the date format to match the espn format
def change_stat_date_format(date):
    string = date

    # extract month and remove from string
    month = date.split(" ", 1)[0]
    string = string.replace(month, "")
    
    # extend the month string to match the espn format
    if "." in month:
        if month == "Jan.":
            month = "January"
        elif month == "Feb.":
            month = "February"
        elif month == "Aug.":
            month = "August"
        elif month == "Sept.":
            month = "September"
        elif month == "Oct.":
            month = "October"
        elif month == "Nov.":
            month = "November"
        elif month == "Dec.":
            month = "December"

    string = month + string
    
    return string
```

```python
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
}

for url in url_list[3500:]:                                # = https://www.fifaindex.com/player/192985/kevin-de-bruyne/
    for date in date_list:                                  # = fifa22_487/
        try:
            # get the whole url
            current_url = url + date

            print("========current: ", current_url, "==============")

            s = requests.Session()
            s.headers['User-Agent'] = headers['User-Agent']
            r = s.get(current_url)
            cur = bs4.BeautifulSoup(r.text, 'html.parser')
            

            # # connect to the url
            # with req.urlopen(current_url) as response:
            #     data = response.read().decode("utf-8-sig")

            # cur=bs4.BeautifulSoup(data, "html.parser")

            # get player's stats
            name = cur.find("li", class_="breadcrumb-item active").getText()
            ID = re.findall(r'\d+', url)[0]
            stat_date = cur.find("a", {'href' : "#", 'data-placement' : "bottom"}).getText()
            stat_date = change_stat_date_format(stat_date)
            height = cur.find("span", {'class' : "data-units data-units-metric"}, string = re.compile("cm$")).getText().replace('cm', '')
            weight = cur.find("span", {'class' : "data-units data-units-metric"}, string = re.compile("kg$")).getText().replace('kg', '')
            if cur.find("span", {'class' : "float-right"}, string = 'Left')!=None :
                preferred_foot = 'Left'
            else:
                preferred_foot = 'Right'
            age = cur.find_all("span", {'class' : "float-right"}, string = re.compile("\d\d"))[1].text
            team = cur.find_all("a", {'class' : "link-team"})[1].text
            number = cur.find_all("span", {'class' : "float-right"})[17].text

            stats = cur.find_all("span", {'class' : re.compile("^badge badge-dark rating")})
            
            # Ball Skills
            ball_control = stats[2].text
            dribbling = stats[3].text

            # Defense
            marking = stats[4].text
            slide_tackle = stats[5].text
            stand_tackle = stats[6].text

            # Mental
            aggression = stats[7].text
            reactions = stats[8].text
            att_position = stats[9].text
            interceptions = stats[10].text
            vision = stats[11].text
            composure = stats[12].text

            # Passing
            crossing = stats[13].text
            short_pass = stats[14].text
            long_pass = stats[15].text

            # Physical
            acceleration = stats[16].text
            stamina = stats[17].text
            strength = stats[18].text
            balance = stats[19].text
            sprint_speed = stats[20].text
            agility = stats[21].text
            jumping = stats[22].text

            # Shooting
            heading = stats[23].text
            shot_power = stats[24].text
            finishing = stats[25].text
            long_shots = stats[26].text
            curve = stats[27].text
            fk_acc = stats[28].text
            penalties = stats[29].text
            volleys = stats[30].text

            # Goalkeeper
            gk_positioning = stats[31].text
            gk_diving = stats[32].text
            gk_handling = stats[33].text
            gk_kicking = stats[34].text
            gk_reflexes = stats[35].text
    
            # Save to csv
            data = ([[  name, ID, stat_date, height, weight, preferred_foot, age, team, number, 
                        ball_control, dribbling, marking, slide_tackle, stand_tackle,
                        aggression, reactions, att_position, interceptions, vision, composure, crossing, short_pass, long_pass, 
                        acceleration, stamina, strength, balance, sprint_speed, agility, jumping, heading, shot_power, finishing, 
                        long_shots, curve, fk_acc, penalties, volleys, gk_positioning, gk_diving, gk_handling, gk_kicking, gk_reflexes]])
    

            file = open('FIFA22_player_stats.csv', 'a+', newline ='', encoding='utf-8-sig')
            
            with file:   
                write = csv.writer(file)
                write.writerows(data)
        except Exception as e:
            print("error on : ", current_url, " error: ", e)
            continue
```
